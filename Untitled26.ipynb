```python
import time
import cProfile
import pstats
import io
import tracemalloc
def swap_numbers(a, b):
    a = a + b
    b = a - b
    a = a - b
    return a, b
def run_timing(a, b):
    print("=== Before Swapping ===")
    print(f"a = {a}, b = {b}\n")

    start = time.perf_counter()
    a, b = swap_numbers(a, b)
    end = time.perf_counter()

    print("=== After Swapping ===")
    print(f"a = {a}, b = {b}\n")
    print("=== Execution Time ===")
    print(f"Time taken : {end - start:.8f} seconds\n")
def run_cpu_profile(a, b):
    profiler = cProfile.Profile()
    profiler.enable()
    swap_numbers(a, b)
    profiler.disable()

    s = io.StringIO()
    stats = pstats.Stats(profiler, stream=s).sort_stats("cumulative")
    stats.print_stats(10)

    print("=== CPU Profiling (Top 10) ===")
    print(s.getvalue())
def run_memory_profile(a, b):
    tracemalloc.start()
    swap_numbers(a, b)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    print("=== Memory Profiling ===")
    print(f"Current memory usage : {current / 10**6:.6f} MB")
    print(f"Peak memory usage    : {peak / 10**6:.6f} MB\n")

def main():
    a = 10
    b = 20
    print("\nSwapping Two Numbers Using Arithmetic Method\n")
    run_timing(a, b)
    run_cpu_profile(a, b)
    run_memory_profile(a, b)

if __name__ == "__main__":
    main()
```
